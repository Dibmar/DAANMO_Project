```python
# All NaN values converted to 0. It is assumed that in this case, NaN signifies either the non-existance of the variable or the nonavailability of the data. New df created to enssuer data integrity.

df_spain_mod = df_spain
df_spain_mod['date'] = pd.to_datetime(df_spain_mod['date'], format='%Y%m%d')
df_spain_mod = df_spain_mod.fillna(0).set_index('date')
df_spain_mod
```

```python
df_spain_mod.isnull().any()
```

```python
# df_spain_mod['total_cases'] = df_spain_mod.new_cases + df_spain_mod.total_cases
df_spain_mod
```

```python
# 10 most contagious days
y_val = df_spain_mod.total_cases.sort_values(ascending = False).head(10)

plt.title('The 10 most contagious days')
plt.ylabel('Number of cases')
y_val.plot(kind= 'bar').legend();
```

```python
y_val=df_spain_mod.total_cases[df_spain_mod.total_cases !=0]

plt.title("Number of cases' evolution")
plt.ylabel('Number of cases')

y_val.plot().legend();
```

```python
# 10 most lethal days
deaths = df_spain_mod['total_deaths'].sort_values(ascending = False).head(10)

plt.title('The 10 most lethal days')
plt.ylabel('Number of deaths')
deaths.plot(kind= 'bar');
```

```python
# Evolution of deaths
y_val=df_spain_mod.total_deaths[df_spain_mod.total_deaths !=0]

plt.title("Number of deaths' evolution")
plt.ylabel('Number of cases')

y_val.plot().legend();
```
